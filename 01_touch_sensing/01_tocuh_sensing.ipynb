```python
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
torch.manual_seed(5)
df = pd.read_excel('touch_sensing_dataset.xlsx')
```

```python
df=df*1
filter = df["touch"]== 1
df.where(filter, inplace = True)
df = df.dropna()
X = df.iloc[:,6:]
```

```python
# For user_id

y = df["user_id"].values # Target
```

```python
class Model(nn.Module):
    def __init__(self, in_features=3200, out_features=30):
        super().__init__()
        self.fc1=nn.Linear(in_features, 1600)
        self.fc2=nn.Linear(1600, 64)
        self.out=nn.Linear(64, out_features)
        
    def forward(self, x):
        x=F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=self.out(x)
        return x
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.25, random_state=5)
```

```python
X_train=torch.FloatTensor(X_train.values)
X_test=torch.FloatTensor(X_test.values)
y_train=torch.LongTensor(y_train)
y_test=torch.LongTensor(y_test) # y_test.values
```

```python
model=Model()
```

```python
criterion=nn.CrossEntropyLoss()  # loss function seçtik
optimizer=torch.optim.SGD(model.parameters(),lr=0.001)  # optim adlı optimizerı seçtik
```

```python
epochs=100
losses=[]
for i in range(epochs):
    y_pred=model.forward(X_train)
    loss=criterion(y_pred, y_train)
    losses.append(loss)
    if i%10==0:
        print(f'Epoch {i} and loss is: {loss}')
        
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
with torch.no_grad():
    y_eval=model.forward(X_test)
    loss=criterion(y_eval, y_test)
```

```python
correct=0
with torch.no_grad():
    for i, data in enumerate(X_test):
        y_val=model.forward(data)
        print(f'{i+1}. {str(y_val)}   {y_test[i]}')
        
        if y_val.argmax().item()==y_test[i]:
            correct+=1
            
print(f'Doğru sayısı: {correct}')
```

```python
percent = (correct/228)*100
percent
```

```python
torch.save(model.state_dict(), "user_id.pt")
```
